### Step 5: Data splits
Every ML project needs three splits:

- **Train**: The model learns from this (typically 80%)
- **Validation**: You check progress during training (typically 10%)
- **Test**: Final evaluation after training is done (typically 10%)


```python
dataset = load_dataset("stanfordnlp/imdb", split="train")
print(dataset)
```

```python
dataset_for_splitting=imdb_dataset["train"]

# seed=42 makes train/test split reproducible by generating same random split every run.
split=dataset_for_splitting.train_test_split(test_size=0.2,shuffle=True,seed=42)

split
```

```python
train_ds=train_val["train"]
val_ds=train_val["test"]
test_ds=split["test"]

"""
At first ypu split in split variable
then you split train_val into train_ds and val_ds
then you split test_ds from split

"""


print(f"Train: {len(train_ds)}, Val: {len(val_ds)}, Test: {len(test_ds)}")
```

#### 2. Stream the c4 dataset and count how many examples you can process in 10 seconds

```python
import time
c4_dataset=load_dataset("allenai/c4", "en",streaming=True,split="train")

start=time.time()
count=0

for i,example in enumerate(c4_dataset):
    count += 1

    if time.time()-start >=10:
        break

end=time.time()
print(f"Time taken: {end-start} seconds")
print(f"Examples processed: {count}")
print(f"Examples per second: {count/(end-start)}")
```

#### 04. Create a 70/15/15 train/val/test split with a fixed seed and verify the sizes


```python
imdb_dataset_ass = datasets.load_dataset("imdb",split="train")
```

```python
split=imdb_dataset_ass.train_test_split(test_size=0.15)

train_val=split['train'].train_test_split(test_size=0.15)

train_dataset = train_val['train']
val_dataset = train_val['test']
test_dataset = split['test']

print(f"Training set size: {len(train_dataset)}")
print(f"Validation set size: {len(val_dataset)}")
print(f"Test set size: {len(test_dataset)}")
```
