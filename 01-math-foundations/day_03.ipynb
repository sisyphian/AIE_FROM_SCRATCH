````python
class Vector:
    def __init__(self,components):
        self.components=list(components)
        self.dim=len(components)

    def __add__(self,other):
        # new_vector=[]
        # for i in range(self.components):
        #     new_vector.append(self.components[i]+other.components[i])

        # return Vector(new_vector)
        return Vector([a+b for a,b in zip(self.components,other.components)])

    def __sub__(self,other):
        # new_vector=[]
        # for i in range(self.components):
        #     new_vector.append(self.components[i]-other.components[i])

        # return Vector(new_vector)
        return Vector([a-b for a,b in zip(self.components,other.components)])

    def dot(self,other):
        # sum=0
        # for i in range(self.components):
        #     sum+=self.components[i]*other.components[i]
        # return sum
        return sum(a*b for a,b in zip(self.components,other.components))
                   
    def magnitude(self):
        # mag_sqaured=0
        # for i in range(self.components):
        #     mag_sqaured+=(self.components[i])**2
        # return math.sqrt(mag_sqaured)
        return sum(x**2 for x in self.components)**0.5

    def normalize(self):
        mag=self.magnitude()
        # normalize_arr=[]

        # for i in range(self.components):
        #     normalize_arr.append(self.components[i]/mag)
        # return Vector(normalize_arr)
        
        return Vector([x/mag for x in self.components])

    def cosine_similarity(self,other):
        # dot_product=self.dot(other)
        # similarity_value=dot_product/(self.magnitude()*other.magnitude())

        # return similarity_value
        return (self.dot(other))/(self.magnitude()*other.magnitude())

    def __repr__(self):
        return f"Vector({self.components})"
    
````

````python
import math


class MyVector:

    def __init__(self, components):
        self.components = list(components)

        # Edge case 1: Empty vector
        if len(self.components) == 0:
            raise ValueError("Vector cannot be empty")

        # Edge case 2: Non-numeric values
        for value in self.components:
            if not isinstance(value, (int, float)):
                raise TypeError("All components must be numeric")

    @property
    def dim(self):
        return len(self.components)

    def _check_vector(self, other):
        # Edge case 3: Wrong operand type
        if not isinstance(other, MyVector):
            raise TypeError("Operand must be MyVector")

        # Edge case 4: Different dimensions
        if self.dim != other.dim:
            raise ValueError("Vectors must have same dimension")

    def __add__(self, other):
        self._check_vector(other)

        # Alternative definition
        # result = []
        # for i in range(self.dim):
        #     result.append(self.components[i] + other.components[i])
        # return MyVector(result)

        return MyVector([
            a + b for a, b in zip(self.components, other.components)
        ])

    def __sub__(self, other):
        self._check_vector(other)

        # Alternative definition
        # result = []
        # for i in range(self.dim):
        #     result.append(self.components[i] - other.components[i])
        # return MyVector(result)

        return MyVector([
            a - b for a, b in zip(self.components, other.components)
        ])

    def dot(self, other):
        self._check_vector(other)

        # Alternative definition
        # total = 0
        # for i in range(self.dim):
        #     total += self.components[i] * other.components[i]
        # return total

        return sum(
            a * b for a, b in zip(self.components, other.components)
        )

    def magnitude(self):

        # Alternative definition
        # mag_squared = 0
        # for value in self.components:
        #     mag_squared += value ** 2
        # return math.sqrt(mag_squared)

        return math.sqrt(
            sum(x ** 2 for x in self.components)
        )

    def normalize(self):
        mag = self.magnitude()

        # Edge case 5: Zero vector
        if mag == 0:
            raise ValueError("Cannot normalize zero vector")

        # Alternative definition
        # normalized = []
        # for value in self.components:
        #     normalized.append(value / mag)
        # return MyVector(normalized)

        return MyVector([
            x / mag for x in self.components
        ])

    def cosine_similarity(self, other):
        self._check_vector(other)

        mag1 = self.magnitude()
        mag2 = other.magnitude()

        # Edge case 6: Zero vector
        if mag1 == 0 or mag2 == 0:
            raise ValueError(
                "Cosine similarity undefined for zero vector"
            )

        # Alternative definition
        # dot_product = self.dot(other)
        # similarity = dot_product / (mag1 * mag2)
        # return similarity

        return self.dot(other) / (mag1 * mag2)

    def __repr__(self):
        return f"MyVector({self.components})"



````

````python
class Matrix:
    def __init__(self, rows):
        self.rows = [list(row) for row in rows]
        self.shape = (len(self.rows), len(self.rows[0]))

    def __matmul__(self, other):
        if isinstance(other, Vector):
            return Vector([
                sum(self.rows[i][j] * other.components[j] for j in range(self.shape[1]))
                for i in range(self.shape[0])
            ])
        rows = []
        for i in range(self.shape[0]):
            row = []
            for j in range(other.shape[1]):
                row.append(sum(
                    self.rows[i][k] * other.rows[k][j]
                    for k in range(self.shape[1])
                ))
            rows.append(row)
        return Matrix(rows)

    def transpose(self):
        return Matrix([
            [self.rows[j][i] for j in range(self.shape[0])]
            for i in range(self.shape[1])
        ])

    def __repr__(self):
        return f"Matrix({self.rows})"


rotation_90 = Matrix([[0, -1], [1, 0]])
point = Vector([3, 1])
print("rotation_90",rotation_90)
rotated = rotation_90 @ point
print(f"Original: {point}")
print(f"Rotated 90°: {rotated}")
````

````python
# Linear independence

def is_linearly_independent(vectors):
    n = len(vectors)
    dim = len(vectors[0].components)
    mat = Matrix([v.components[:] for v in vectors])
    rows = [row[:] for row in mat.rows]
    rank = 0
    for col in range(dim):
        pivot = None
        for row in range(rank, len(rows)):
            if abs(rows[row][col]) > 1e-10:
                pivot = row
                break
        if pivot is None:
            continue
        rows[rank], rows[pivot] = rows[pivot], rows[rank]
        scale = rows[rank][col]
        rows[rank] = [x / scale for x in rows[rank]]
        for row in range(len(rows)):
            if row != rank and abs(rows[row][col]) > 1e-10:
                factor = rows[row][col]
                rows[row] = [rows[row][j] - factor * rows[rank][j] for j in range(dim)]
        rank += 1
    return rank == n


def project(a, b):
    scalar = a.dot(b) / b.dot(b)
    return Vector([scalar * x for x in b.components])


def gram_schmidt(vectors):
    orthonormal = []
    for v in vectors:
        w = v
        for u in orthonormal:
            proj = project(w, u)
            w = w - proj
        if w.magnitude() < 1e-10:
            continue
        orthonormal.append(w.normalize())
    return orthonormal


v1 = Vector([1, 0, 0])
v2 = Vector([1, 1, 0])
v3 = Vector([1, 1, 1])
basis = gram_schmidt([v1, v2, v3])
for i, u in enumerate(basis):
    print(f"u{i+1} = {u}")
    print(f"  |u{i+1}| = {u.magnitude():.6f}")

print(f"u1 · u2 = {basis[0].dot(basis[1]):.6f}")
print(f"u1 · u3 = {basis[0].dot(basis[2]):.6f}")
print(f"u2 · u3 = {basis[1].dot(basis[2]):.6f}")
````

````python
import math


class Matrix:

    def __init__(self, rows):

        if not rows:
            raise ValueError("Matrix cannot be empty")

        self.rows = [list(row) for row in rows]

        cols = len(self.rows[0])

        for row in self.rows:
            if len(row) != cols:
                raise ValueError("All rows must have same length")

        self.shape = (len(self.rows), cols)

    def __add__(self, other):

        if not isinstance(other, Matrix):
            return NotImplemented

        if self.shape != other.shape:
            raise ValueError("Matrices must have same shape")

        result = []

        for i in range(self.shape[0]):

            row = []

            for j in range(self.shape[1]):
                row.append(self.rows[i][j] + other.rows[i][j])

            result.append(row)

        return Matrix(result)

    def __sub__(self, other):

        if not isinstance(other, Matrix):
            return NotImplemented

        if self.shape != other.shape:
            raise ValueError("Matrices must have same shape")

        result = []

        for i in range(self.shape[0]):

            row = []

            for j in range(self.shape[1]):
                row.append(self.rows[i][j] - other.rows[i][j])

            result.append(row)

        return Matrix(result)

    def scalar_multiply(self, number):

        if not isinstance(number, (int, float)):
            return NotImplemented

        result = []

        for row in self.rows:
            result.append([number * x for x in row])

        return Matrix(result)

    def element_wise_multiply(self, other):

        if not isinstance(other, Matrix):
            return NotImplemented

        if self.shape != other.shape:
            raise ValueError("Matrices must have same shape")

        result = []

        for i in range(self.shape[0]):

            row = []

            for j in range(self.shape[1]):
                row.append(self.rows[i][j] * other.rows[i][j])

            result.append(row)

        return Matrix(result)

    def __matmul__(self, other):

        if isinstance(other, Vector):

            if self.shape[1] != other.dim:
                raise ValueError("Incompatible dimensions")

            return Vector([
                sum(
                    self.rows[i][j] * other.components[j]
                    for j in range(self.shape[1])
                )
                for i in range(self.shape[0])
            ])

        if not isinstance(other, Matrix):
            return NotImplemented

        if self.shape[1] != other.shape[0]:
            raise ValueError("Incompatible matrix dimensions")

        result = []

        for i in range(self.shape[0]):

            row = []

            for j in range(other.shape[1]):

                value = sum(
                    self.rows[i][k] * other.rows[k][j]
                    for k in range(self.shape[1])
                )

                row.append(value)

            result.append(row)

        return Matrix(result)

    def transpose(self):

        result = []

        for i in range(self.shape[1]):

            row = []

            for j in range(self.shape[0]):
                row.append(self.rows[j][i])

            result.append(row)

        return Matrix(result)

    def __repr__(self):

        rows = "\n  ".join(str(row) for row in self.rows)

        return f"Matrix({self.shape}):\n  {rows}"
````

In neural networks, weight matrices transform input vectors into output vectors. A layer with 784 inputs and 128 outputs uses a 128x784 weight matrix.

Matrix multiplication has a strict rule: (m x n) @ (n x p) = (m x p). The inner dimensions must match.

```
(128 x 784) @ (784 x 1) = (128 x 1)
  weights       input       output

Inner dimensions: 784 = 784  -- valid
```

Operation	What it does	Neural network use
Addition	Element-wise combine	Adding bias to output
Scalar multiply	Scale every element	Learning rate * gradients
Matrix multiply	Transform vectors	Layer forward pass
Transpose	Flip rows and columns	Backpropagation
Determinant	Single number summary	Checking invertibility
Inverse	Undo a transformation	Solving linear systems
Identity	Do-nothing matrix	Initialization, residual connections

### Element-wise vs matrix multiplication

1. Element-wise: multiply matching positions. Both matrices must be the same shape.
```
| 1  2 |   | 5  6 |   | 5  12 |
| 3  4 | * | 7  8 | = | 21 32 |
```

2. Matrix multiplication: dot products of rows and columns. Inner dimensions must match.
```
| 1  2 |   | 5  6 |   | 1*5+2*7  1*6+2*8 |   | 19  22 |
| 3  4 | @ | 7  8 | = | 3*5+4*7  3*6+4*8 | = | 43  50 |
```

### Broadcasting
When you add a bias vector to a matrix of outputs, the shapes do not match. Broadcasting stretches the smaller array to fit.

```
| 1  2  3 |   +   [10, 20, 30]
| 4  5  6 |

Broadcasting stretches the vector across rows:

| 1  2  3 |   | 10  20  30 |   | 11  22  33 |
| 4  5  6 | + | 10  20  30 | = | 14  25  36 |
```


````python
# latest matrix implementation
class Matrix:
    def __init__(self, data):
        self.data = [list(row) for row in data]
        self.rows = len(self.data)
        self.cols = len(self.data[0])
        self.shape = (self.rows, self.cols)

    def __repr__(self):
        rows_str = "\n  ".join(str(row) for row in self.data)
        return f"Matrix({self.shape}):\n  {rows_str}"

    def __add__(self, other):
        return Matrix([
            [self.data[i][j] + other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def __sub__(self, other):
        return Matrix([
            [self.data[i][j] - other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def scalar_multiply(self, scalar):
        return Matrix([
            [self.data[i][j] * scalar for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def element_wise_multiply(self, other):
        return Matrix([
            [self.data[i][j] * other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def matmul(self, other):
        return Matrix([
            [
                sum(self.data[i][k] * other.data[k][j] for k in range(self.cols))
                for j in range(other.cols)
            ]
            for i in range(self.rows)
        ])

    def transpose(self):
        return Matrix([
            [self.data[j][i] for j in range(self.rows)]
            for i in range(self.cols)
        ])

    def determinant(self):
        if self.shape == (1, 1):
            return self.data[0][0]
        if self.shape == (2, 2):
            return self.data[0][0] * self.data[1][1] - self.data[0][1] * self.data[1][0]
        det = 0
        for j in range(self.cols):
            minor = Matrix([
                [self.data[i][k] for k in range(self.cols) if k != j]
                for i in range(1, self.rows)
            ])
            det += ((-1) ** j) * self.data[0][j] * minor.determinant()
        return det

    def inverse_2x2(self):
        det = self.determinant()
        if det == 0:
            raise ValueError("Matrix is singular, no inverse exists")
        return Matrix([
            [self.data[1][1] / det, -self.data[0][1] / det],
            [-self.data[1][0] / det, self.data[0][0] / det]
        ])


    @staticmethod
    def identity(n):
        return Matrix([
            [1 if i == j else 0 for j in range(n)]
            for i in range(n)
        ])
````

````python
# latest matrix implementation
class Matrix:
    def __init__(self, data):
        self.data = [list(row) for row in data]
        self.rows = len(self.data)
        self.cols = len(self.data[0])
        self.shape = (self.rows, self.cols)

    def __repr__(self):
        rows_str = "\n  ".join(str(row) for row in self.data)
        return f"Matrix({self.shape}):\n  {rows_str}"

    def __add__(self, other):
        return Matrix([
            [self.data[i][j] + other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def __sub__(self, other):
        return Matrix([
            [self.data[i][j] - other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def scalar_multiply(self, scalar):
        return Matrix([
            [self.data[i][j] * scalar for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def element_wise_multiply(self, other):
        return Matrix([
            [self.data[i][j] * other.data[i][j] for j in range(self.cols)]
            for i in range(self.rows)
        ])

    def matmul(self, other):
        return Matrix([
            [
                sum(self.data[i][k] * other.data[k][j] for k in range(self.cols))
                for j in range(other.cols)
            ]
            for i in range(self.rows)
        ])

    def transpose(self):
        return Matrix([
            [self.data[j][i] for j in range(self.rows)]
            for i in range(self.cols)
        ])

    def determinant(self):
        if self.shape == (1, 1):
            return self.data[0][0]
        if self.shape == (2, 2):
            return self.data[0][0] * self.data[1][1] - self.data[0][1] * self.data[1][0]
        det = 0
        for j in range(self.cols):
            minor = Matrix([
                [self.data[i][k] for k in range(self.cols) if k != j]
                for i in range(1, self.rows)
            ])
            det += ((-1) ** j) * self.data[0][j] * minor.determinant()
        return det

    def inverse_2x2(self):
        det = self.determinant()
        if det == 0:
            raise ValueError("Matrix is singular, no inverse exists")
        return Matrix([
            [self.data[1][1] / det, -self.data[0][1] / det],
            [-self.data[1][0] / det, self.data[0][0] / det]
        ])


    @staticmethod
    def identity(n):
        return Matrix([
            [1 if i == j else 0 for j in range(n)]
            for i in range(n)
        ])
````

## 03 Matrix Transformations (need deepr understanding)

```
A @ v = lambda * v

v is the eigenvector (direction that survives)
lambda is the eigenvalue (how much it stretches)

Example: A = | 2  1 |
             | 1  2 |

Eigenvector [1, 1] with eigenvalue 3:
  A @ [1,1] = [3, 3] = 3 * [1, 1]     (same direction, scaled by 3)

Eigenvector [1, -1] with eigenvalue 1:
  A @ [1,-1] = [1, -1] = 1 * [1, -1]  (same direction, unchanged)

  A = V @ D @ V^(-1)

V = matrix whose columns are eigenvectors
D = diagonal matrix of eigenvalues
V^(-1) = inverse of V

This says: rotate into eigenvector coordinates, scale along each axis, rotate back.

  ```

Think about people.

Need 3 people move table.

### Case 1: Linearly Independent

Person A pushes east.

→

Person B pushes north.

↑

Person C pushes upward.

Out of screen.

All three do different jobs.

Remove any one person.

Cannot replace them using other two.

This is **linearly independent**.

They can reach any point in 3D space.

---

### Case 2: Linearly Dependent

Person A:

→

Person B:

↑

Person C:

↗

But

```
Person C = Person A + Person B
```

Third person adds nothing new.

Already can make C by combining A and B.

So only 2 independent directions exist.

Span becomes plane instead of whole 3D space.

---

Simple vector example

```
v1 = (1, 0)
v2 = (0, 1)
```

Independent.

Cannot make `(1,0)` from `(0,1)`.

Cannot make `(0,1)` from `(1,0)`.

Now add

```
v3 = (1, 1)
```

Notice

```
v3 = v1 + v2
```

So `v3` not new information.

Dependent.

---

AI example

Suppose predicting house price.

Features:

```
Feature 1: Height in meters
Feature 2: Weight in kg
Feature 3: Age
```

Good.

Each tells something different.

Now imagine

```
Feature 1: Height in cm
Feature 2: Height in meters
```

But

```
Height(cm) = 100 × Height(m)
```

Same information.

One feature completely determined by other.

Model asks:

"Should weight go to Feature 1 or Feature 2?"

No unique answer.

Maybe

```
Weight1 = 5
Weight2 = -4
```

or

```
Weight1 = 100
Weight2 = -99
```

Both produce nearly same prediction.

Weights become unstable.

Tiny data change → huge weight change.

This called **multicollinearity**.

---

Easy rule:

* Independent = each feature adds new information.
* Dependent = one feature can be built from others.
* AI wants independent features because model learns stable, meaningful weights.
