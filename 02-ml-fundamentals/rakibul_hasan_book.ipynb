````python
x_mean=sum(X)/len(X)
y_mean=sum(y)/len(y)
y_mean
````

````python
# m=Sum_of_all(x-x_mean)*(y-y_mean)/Sum_of(x-x_mean)**2
````

````python
x_diff=[]
for x in X:
    x_diff.append(x-x_mean)
    print(f"x: {x}-{x_mean} =  {x-x_mean}")
````

````python
y_diff=[]
for y_i in y:
    y_diff.append(y_i-y_mean)
    print(f"y: {y_i}-{y_mean} =  {y_i-y_mean}")
````

````python
x_diff
````

````python
y_diff
````

````python
# upper portion 
upper_sum=0

for i , j in zip(x_diff,y_diff):
    upper_sum+=i*j
    print(f"{i}*{j} = {i*j}")
````

````python
upper_sum
````

````python
lower_sum=0
for i in x_diff:
    lower_sum+=i**2
lower_sum
````

````python
m=upper_sum/lower_sum
m
````

````python
b=y_mean-m*x_mean
b
````

````python
def model(x):
    return m*x+b
````

````python
y_predicted=[]
x_for_plot=[1,2,3,4,5]

for x in x_for_plot:
    y_predicted.append(model(x))
    print(f"for {x} prediction is {model(x)}")
````

````python
import matplotlib.pyplot as plt

plt.xlabel("Month")
plt.ylabel("N of Video")
plt.grid(True)
plt.xticks(x_for_plot)
plt.yticks(y_predicted)
plt.plot([2,3,4,6],to_plot)
plt.plot(X,y)
plt.plot(x_for_plot,y_predicted , )
````

````python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 5))

# Actual data points
plt.plot(X, y, color="blue", label="Actual Data")

# Highlight the point(s) you want
plt.plot([2, 3, 4, 6], to_plot,
            color="red", label="previous line with default tangent calculation ")

# Regression line
plt.plot(x_for_plot, y_predicted,
         color="green", linewidth=2.5,
         label="Least Squares Regression")

# Labels and title
plt.title("Least Squares Regression")
plt.xlabel("Month")
plt.ylabel("Number of Videos")

# Grid
plt.grid(True, linestyle="--", alpha=0.6)

# Nice tick spacing
plt.xticks(np.arange(min(x_for_plot), max(x_for_plot) + 1, 1))

plt.yticks(
    np.arange(
        0,
        max(max(y), max(y_predicted)) + 5,
        5
    )
)

# Legend
plt.legend()

# Prevent labels from being cut off
plt.tight_layout()

plt.show()
````
