```python
def predict(x, m, c):
    return m * x + c


def calculate_gradient(X, y, m, c):
    n = len(X)

    dw = 0
    db = 0

    for x, y_i in zip(X, y):
        y_hat = predict(x, m, c)

        dw += (y_hat - y_i) * x
        db += (y_hat - y_i)

    dw = (2 / n) * dw
    db = (2 / n) * db

    return {"dw": dw, "db": db}
```

```python
# lets take a single gradient for value m 
grad_for_0=calculate_gradient(X,y,3,c)
grad_for_0
```

```python
dws=[]
dbs=[]

for m in value_of_m:
    grad=calculate_gradient(X,y,m,0)
    dws.append(grad["dw"])
    dbs.append(grad["db"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Sample Data
# -----------------------------
X = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

c = 0  # keep intercept fixed


# -----------------------------
# Prediction
# -----------------------------
def predict(x, m, c):
    return m * x + c


# -----------------------------
# MSE
# -----------------------------
def mse(X, y, m, c):
    y_pred = predict(X, m, c)
    return np.mean((y_pred - y) ** 2)


# -----------------------------
# Gradient wrt m
# -----------------------------
def gradient_m(X, y, m, c):
    y_pred = predict(X, m, c)
    return (2 / len(X)) * np.sum((y_pred - y) * X)


# -----------------------------
# Cost Curve
# -----------------------------
m_values = np.linspace(-1, 5, 300)
cost_values = [mse(X, y, m, c) for m in m_values]

plt.figure(figsize=(9, 6))
plt.plot(m_values, cost_values, linewidth=2, label="Cost Curve")

# -----------------------------
# Choose a point
# -----------------------------
m = 1.5

cost = mse(X, y, m, c)
dw = gradient_m(X, y, m, c)

# Draw current point
plt.scatter(m, cost, color="red", s=80, zorder=5)

# -----------------------------
# Tangent
# -----------------------------
dx = 0.02  # short segment

x = np.linspace(m - dx, m + dx, 100)
tangent = cost + dw * (x - m)

plt.plot(x, tangent, "g--", linewidth=3, label="Tangent")

# -----------------------------
# Beautify
# -----------------------------
plt.xlabel("m")
plt.ylabel("Cost (MSE)")
plt.title("Gradient Descent Tangent Visualization")
plt.grid(True)
plt.legend()

# Keep the cost curve visible
plt.xlim(min(m_values), max(m_values))
plt.ylim(min(cost_values) - 5, max(cost_values) + 5)

plt.show()
```
