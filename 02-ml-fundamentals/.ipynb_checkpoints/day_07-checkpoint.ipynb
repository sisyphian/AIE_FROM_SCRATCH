#### The Bias-Variance Tradeoff
This is the mathematical framework behind overfitting and underfitting.

**Bias**: Error from wrong assumptions in the model. A linear model has high bias when the true relationship is nonlinear. High bias leads to underfitting.

**Variance**: Error from sensitivity to small fluctuations in the training data. A model with high variance gives very different predictions when trained on different subsets of data. High variance leads to overfitting.

### Implementing Nearest Centroid classifier

````python
import numpy as np
````

````python
class NearestCentroid:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.centroids = np.array([
            X[y == c].mean(axis=0) for c in self.classes
        ])

    def predict(self, X):
        distances = np.array([
            np.sqrt(((X - c) ** 2).sum(axis=1))
            for c in self.centroids
        ])
        return self.classes[distances.argmin(axis=0)]
````

03. **A model gets 99% accuracy on training data but 60% on test data. Diagnose the problem and list three things you would try to fix it.**

Diagnosis

```text
Training Accuracy = 99%
Test Accuracy = 60%
```

Large gap.

Problem = **Overfitting**.

Model learned training data too well, including noise and random patterns.

Result

* Excellent on training data.
* Poor on unseen test data.
* Generalization weak.

Three fixes

### 1. Reduce model complexity

Model too complex.

Examples

* Smaller neural network.
* Lower `max_depth` in Decision Tree.
* Fewer parameters.
* Prune tree.

Less complexity → less overfitting.

---

### 2. Add regularization

Penalize overly complex model.

Examples

* L1 regularization (Lasso)
* L2 regularization (Ridge)
* Dropout (Neural Networks)
* Weight decay

Model focuses important patterns instead of memorizing.

---

### 3. More training data

More diverse examples.

Harder for model memorize every sample.

Model learns general patterns instead of noise.

Other valid fixes

* Early stopping.
* Data augmentation.
* Cross-validation.
* Feature selection.
* Tune hyperparameters.
* Remove noisy features.
* Collect cleaner data.
