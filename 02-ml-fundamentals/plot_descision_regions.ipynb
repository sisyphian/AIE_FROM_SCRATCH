```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
```

```python
# Step 1: Create dataset
X, y = make_blobs(
    n_samples=100,
    centers=2,
    random_state=42,
    cluster_std=1.5
)
```

```python
X
```

```python
y
```

```python
model = SVC(kernel="linear")
```

```python
model.fit(X,y)
```

```python
# --------------------------------------------------
# Step 3: Find plotting range
# --------------------------------------------------
x_min=X[:,0].min() -1 # adding padding for extra width in space
x_max = X[:,0].max() +1 # adding padding for extra width in space

y_min=X[:,1].min() -1 # adding padding for extra width in space
y_max = X[:,1].max() +1 # adding padding for extra width in space
```

```python
# --------------------------------------------------
# Step 4: Create meshgrid
# xx -> x coordinates
# yy -> y coordinates
# --------------------------------------------------
xx,yy=np.meshgrid(
    np.linspace(x_min,x_max),
    np.linspace(y_min,y_max),
)
```

```python
xx
```

```python
grid=np.c_[xx.ravel(),yy.ravel()]
```

```python
grid
```

```python
Z = model.predict(grid)
```

```python
Z
```

```python
Z=Z.reshape(xx.shape)
Z
```

```python
plt.contourf(xx,yy,Z,alpha=0.3 , cmap="bwr")
```

```python
# plotting original data 
plt.scatter(X[:,0],X[:,1],c=y , edgecolors="k")
plt.contourf(xx,yy,Z,alpha=0.3 , cmap="bwr")
```
